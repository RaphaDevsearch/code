```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python

def visualize_graph(graph, title="Social Network Graph"):
  """
  Visualize a graph dictionary.
  graph:
  {
      node: {neighbors}
  }
  """
  # Create graph object
  G = nx.Graph()
  # Add nodes and edges
  for node, neighbors in graph.items():
    for neighbor in neighbors:
      G.add_edge(node, neighbor)
  # Create positions
  pos = nx.spring_layout(
    G,
    seed=42
  )
  plt.figure(figsize=(8,6))
  # Draw edges
  nx.draw_networkx_edges(
    G,
    pos,
    width=2
  )
  # Draw nodes
  nx.draw_networkx_nodes(
    G,
    pos,
    node_size=1500
  )
  # Draw labels
  nx.draw_networkx_labels(
    G,
    pos,
    font_size=12,
    font_weight="bold"
  )
  plt.title(title)
  plt.axis("off")
  plt.show()
```

```python
"""
Graph representes relationship between entities (NODES),
A graph is a data structure in DSA that models :
  - Objects/entities = NODE (vertices)
  - relationships/connections = edge
"""
graph = {
  "Raphael": {"John", "Sarah"},
  "John": {"Raphael", "Mike", "Emma"},
  "Sarah": {"Raphael", "David", "Lisa"},
  "Mike": {"John"},
  "Emma": {"John"},
  "David": {"Sarah"},
  "Lisa": {"Sarah","Raphael"}
}

visualize_graph(graph)
```
